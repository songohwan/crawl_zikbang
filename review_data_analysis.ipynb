```python
from kiwipiepy import Kiwi
```

```python
kiwi = Kiwi()
```

```python
korean_stop_words = 'korean_stopwords.txt'

with open(korean_stop_words, encoding='utf-8') as f:
    final_stopword = f.readlines()
final_stopword = [x.strip() for x in final_stopword]
print(final_stopword[:10])
```

```python
stopwords = ['도', '는', '다', '의', '가', '이', '은', '한', '에', '하', '고', '을', '를', '인', '듯', '과', '와', '네', '들', '듯', '지', '임', '게', '이라']

for i in stopwords:
    final_stopword.append(i)
```

```python
def tokenizing(sentence):
    words = [token[0] for token in kiwi.tokenize(sentence) if token[0] not in final_stopword]
    return words
```

```python
review_tokenized = review['transportaion_review'].apply(lambda x: tokenizing(x))
```

```python
appended_list = []
'''Apply 사용해서 List를 하나로 합치는 방법 없음?'''
for i in range(len(review_tokenized)):
    for x in range(len(review_tokenized[i])):
        appended_list.append(review_tokenized[i][x])
```

```python
import re
```

```python
appended_list = [re.sub(r'\n', '', x) for x in appended_list]
```

```python
from collections import Counter
```

```python
counter = Counter(appended_list)
```

```python
remove_one_word = Counter({x: counter[x] for x in counter if len(x) > 1})
```

```python
print(remove_one_word)
```

```python
import random
import pytagcloud
import webbrowser
```

```python
ranked_tags = remove_one_word.most_common(40)
```

```python
tag_list = pytagcloud.make_tags(ranked_tags, maxsize = 80)
```

```python
pytagcloud.create_tag_image(tag_list, 'wordcloud.jpg', size = (900, 600), fontname='NanumGothic', rectangular=False)
```

```python
pytagcloud.create_tag_image(tag_list, 'transportation.jpg', size = (900, 600), fontname='NanumGothic', rectangular=False)
```
